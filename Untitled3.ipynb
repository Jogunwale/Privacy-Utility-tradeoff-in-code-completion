```python
#imports
import re
import torch
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
dataset = load_dataset("openai/openai_humaneval")
#selecting the first 20 examples
test_dataset = dataset["test"]
first_20 = test_dataset.select(range(20))
example = first_20[0]
example
```

```python
def low_obfuscation(prompt: str) -> str:
    #This will rename variable and argument names but preserve structure and comments
    arg_pattern = r"def\s+\w+\((.*?)\):"
    match = re.search(arg_pattern, prompt)

    if not match:
        return prompt

    args = match.group(1)
    arg_names = [
        a.strip().split("=")[0]
        for a in args.split(",")
        if a.strip()
    ]

    rename_map = {name: f"var{i}" for i, name in enumerate(arg_names)}

    obfuscated = prompt
    for original, replacement in rename_map.items():
        obfuscated = re.sub(rf"\b{original}\b", replacement, obfuscated)

    return obfuscated
```

```python
def high_obfuscation(prompt: str) -> str:
    #This is to strip comments, docstrings and replace identifiers with placeholders
    # Remove docstrings
    obfuscated = re.sub(r'""".*?"""', '', prompt, flags=re.DOTALL)
    obfuscated = re.sub(r"'''.*?'''", '', obfuscated, flags=re.DOTALL)

    # Remove single line comments
    obfuscated = re.sub(r"#.*", "", obfuscated)

    # Rename function and arguments
    func_pattern = r"def\s+(\w+)\((.*?)\):"
    match = re.search(func_pattern, obfuscated)

    if not match:
        return obfuscated.strip()

    func_name, args = match.groups()
    arg_names = [
        a.strip().split("=")[0]
        for a in args.split(",")
        if a.strip()
    ]

    rename_map = {func_name: "f"}
    rename_map.update({name: f"x{i}" for i, name in enumerate(arg_names)})

    for original, replacement in rename_map.items():
        obfuscated = re.sub(rf"\b{original}\b", replacement, obfuscated)

    return obfuscated.strip()
```

```python
#load codegen tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("Salesforce/codegen-350M-mono")
model = AutoModelForCausalLM.from_pretrained("Salesforce/codegen-350M-mono")
#set pad token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = AutoModelForCausalLM.from_pretrained("Salesforce/codegen-350M-mono").to(device)
#generation function
def generate_completion(prompt, max_new_tokens=256):
    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True
    )
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=tokenizer.pad_token_id
        )
    # Extract only the generated part (not the input prompt)
    full_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    completion = full_text[len(prompt):]
    return completion
```

```python
results = []
for idx, example in enumerate(first_20):
    print(f"Processing example {idx + 1}/20...")
    original = example["prompt"]
    low = low_obfuscation(original)
    high = high_obfuscation(original)
    for level, prompt in [
        ("original", original),
        ("low", low),
        ("high", high),
    ]:
        print(f"  Generating {level}...")
        completion = generate_completion(prompt)
        results.append({
            "task_id": example["task_id"],
            "obfuscation": level,
            "prompt": prompt,
            "completion": completion,
            "canonical_solution": example["canonical_solution"],
        })
print(f"Done! Generated {len(results)} completions")
```

```python
#install evaluation libraries
!pip install rouge-score python-Levenshtein

from rouge_score import rouge_scorer
import Levenshtein
```

```python
#evaluation functions
def calculate_utility(completion: str, canonical: str) -> float:
    """Calculate utility using ROUGE-L F1 score"""
    scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=False)
    scores = scorer.score(canonical, completion)
    return scores['rougeL'].fmeasure

def calculate_privacy(obfuscated_prompt: str, original_prompt: str) -> float:
    """
    Calculate privacy as normalized Levenshtein distance.
    Higher score = more privacy (more different from original)
    """
    distance = Levenshtein.distance(obfuscated_prompt, original_prompt)
    max_len = max(len(obfuscated_prompt), len(original_prompt))
    if max_len == 0:
        return 0.0
    normalized_distance = distance / max_len
    return normalized_distance
```

```python
# Store original prompts for privacy calculation
original_prompts = {example["task_id"]: example["prompt"] for example in first_20}

# Evaluate all 60 results
for result in results:
    # Calculate utility score
    result["utility_score"] = calculate_utility(
        result["completion"],
        result["canonical_solution"]
    )
    # Calculate privacy score
    original_prompt = original_prompts[result["task_id"]]
    result["privacy_score"] = calculate_privacy(
        result["prompt"],
        original_prompt
    )
```

```python
# Display evaluation results
print("Evaluation complete!")
print(f"\nExample result:")
print(f"Task ID: {results[0]['task_id']}")
print(f"Obfuscation: {results[0]['obfuscation']}")
print(f"Utility Score: {results[0]['utility_score']:.4f}")
print(f"Privacy Score: {results[0]['privacy_score']:.4f}")
```

```python
#create scatter plot
import matplotlib.pyplot as plt

# Separate by obfuscation level
original_results = [r for r in results if r["obfuscation"] == "original"]
low_results = [r for r in results if r["obfuscation"] == "low"]
high_results = [r for r in results if r["obfuscation"] == "high"]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 7))

ax.scatter(
    [r["privacy_score"] for r in original_results],
    [r["utility_score"] for r in original_results],
    label="Original", alpha=0.6, s=100, marker='o'
)
ax.scatter(
    [r["privacy_score"] for r in low_results],
    [r["utility_score"] for r in low_results],
    label="Low Obfuscation", alpha=0.6, s=100, marker='s'
)
ax.scatter(
    [r["privacy_score"] for r in high_results],
    [r["utility_score"] for r in high_results],
    label="High Obfuscation", alpha=0.6, s=100, marker='^'
)

ax.set_xlabel("Privacy Score (Normalized Levenshtein Distance)", fontsize=10)
ax.set_ylabel("Utility Score (ROUGE-L F1)", fontsize=10)
ax.set_title("Privacy-Utility Trade-off in Code Completion", fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
# Calculate average scores by obfuscation level
for level in ["original", "low", "high"]:
    level_results = [r for r in results if r["obfuscation"] == level]
    avg_privacy = np.mean([r["privacy_score"] for r in level_results])
    avg_utility = np.mean([r["utility_score"] for r in level_results])
    print(f"{level.upper()}:")
    print(f"  Average Privacy Score: {avg_privacy:.4f}")
    print(f"  Average Utility Score: {avg_utility:.4f}\n")
```
